데이터 규모

```python
csvpath = os.path.join(drivepath, "1/")
tacsv = os.path.join(csvpath, "training_acc.csv")
tlcsv = os.path.join(csvpath, "training_loss.csv")
vacsv = os.path.join(csvpath, "validation_acc.csv")
vlcsv = os.path.join(csvpath, "validation_loss.csv")
```

```python
df = pd.read_csv(tacsv)

LEN = 268
X = list(df["Step"])[:LEN]
va1 = list(df["silvery-brook-1 - training_acc"])[:LEN]
va2 = list(df["apricot-night-3 - training_acc"])[:LEN]
va3 = []
for idx in range(len(df["leafy-forest-16 - training_acc"])):
  if idx % 2 == 0:
    va3.append(df["leafy-forest-16 - training_acc"][idx])
va3 = va3[:LEN]

plt.title("Training Accuracy")
plt.plot(X, va1, label="1st")
plt.plot(X, va2, color="green", label = "2nd")
plt.plot(X, va3, color="firebrick", label = '3rd')
plt.xticks([0, 54, 108, 162, 217, 271], (0, 1, 2, 3, 4, 5))
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
df = pd.read_csv(vacsv)

X = list(df["Step"])
a1 = list(df["silvery-brook-1 - validation_acc"])
a2 = list(df["apricot-night-3 - validation_acc"])
a3 = list(df["leafy-forest-16 - validation_acc"])

plt.title("Validation Accuracy")
plt.plot(X, a1, label="1st")
plt.plot(X, a2, color="green", label = "2nd")
plt.plot(X, a3, color="firebrick", label = '3rd')
plt.xticks([0, 54, 108, 162, 217, 271], (0, 1, 2, 3, 4, 5))
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
df = pd.read_csv(tlcsv)

LEN = 268
X = list(df["Step"])[:LEN]
va1 = list(df["silvery-brook-1 - training_loss"])[:LEN]
va2 = list(df["apricot-night-3 - training_loss"])[:LEN]
va3 = []
for idx in range(len(df["leafy-forest-16 - training_loss"])):
  if idx % 2 == 0:
    va3.append(df["leafy-forest-16 - training_loss"][idx])
va3 = va3[:LEN]

plt.title("Training Loss")
plt.plot(X, va1, label="1st")
plt.plot(X, va2, color="green", label = "2nd")
plt.plot(X, va3, color="firebrick", label = '3rd')
plt.xticks([0, 54, 108, 162, 217, 271], (0, 1, 2, 3, 4, 5))
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
df = pd.read_csv(vlcsv)

X = list(df["Step"])
a1 = list(df["silvery-brook-1 - validation_loss"])
a2 = list(df["apricot-night-3 - validation_loss"])
a3 = list(df["leafy-forest-16 - validation_loss"])

plt.title("Validation Loss")
plt.plot(X, a1, label="1st")
plt.plot(X, a2, color="green", label = "2nd")
plt.plot(X, a3, color="firebrick", label = '3rd')
plt.xticks([0, 54, 108, 162, 217, 271], (0, 1, 2, 3, 4, 5))
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
