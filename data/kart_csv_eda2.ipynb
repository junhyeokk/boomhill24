```python
import os
from typing import Dict
from collections import defaultdict
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 64가지 경우의 수에 대한 분포

def CsvEda(csvFile):

  print(csvFile, '\n')

  df = pd.read_csv(csvFile, names=["filename", "input"], dtype={"filename": str, "input": str})

  # init dict
  key_counts = {}
  for num in range(64):
    k = bin(num)[2:]    # 000000 부터 111111 까지
    k = k.zfill(6)
    key_counts[k] = []

  for input in df["input"]:
    for k in key_counts.keys():
      key_counts[k].append(0)
    key_counts[input][-1] = 1    # 눌려진 키만 1로

  plt.rcParams['figure.figsize'] = [20, 64]
  fig = plt.figure()

  ax = []
  for idx in range(64):
    ax.append(fig.add_subplot(64, 1, idx+1))
    #ax[idx].axes.xaxis.set_visible(False)  # ticks  label 다 제거
    #ax[idx].axes.yaxis.set_visible(False)
    ax[idx].axes.xaxis.set_ticklabels([])  # ticks 제거
    ax[idx].axes.yaxis.set_ticklabels([])
    k = bin(idx)[2:]
    k = k.zfill(6)
    ax[idx].bar(range(len(df)), key_counts[k], width=1.0)
    ax[idx].set_ylabel(k)
    ax[idx].set_xlim(0, len(df))                 # x축 범위
    ax[idx].set_ylim(0, 1)                       # y축 범위

  plt.show()

  key_total = np.zeros(64)
  key_counts_list = list(key_counts.values())
  for idx in range(64):
    key_total[idx] = sum(key_counts_list[idx])

  print(df["input"].value_counts())

  plt.rcParams['figure.figsize'] = [16, 6]
  fig = plt.figure()
  plt.bar(key_counts.keys(), key_total)
  plt.xticks(rotation=270)
  plt.show()

  print("")
```

```python
drivepath = "/content/gdrive/MyDrive/kart/"
csvpath = os.path.join(drivepath, "csv/")  # csv 파일 경로
csvList = sorted( os.listdir(csvpath) )
print(csvList)
```

```python
for csvFile in csvList:
  csvFile = os.path.join(csvpath, csvFile)
  CsvEda(csvFile)
```

```python
df = pd.DataFrame()

for csvFile in csvList:
  csvFile = os.path.join(csvpath, csvFile)
  temp = pd.read_csv(csvFile, names=["filename", "input"], dtype={"filename": str, "input": str})
  df = pd.concat((df, temp))

labels = ["up", "left", "right", "ctrl", "shift", "down"]
key_counts = [[], [], [], [], [], []]

for input in df["input"]:
  for idx in range(6):
    if '1' == input[idx]:
      key_counts[idx].append(1)
    else:
      key_counts[idx].append(0)

plt.rcParams['figure.figsize'] = [20,6]
fig = plt.figure()

ax = []
for idx in range(6):
  ax.append(fig.add_subplot(6, 1, idx+1))
  #ax[idx].axes.xaxis.set_visible(False)  # ticks  label 다 제거
  #ax[idx].axes.yaxis.set_visible(False)
  ax[idx].axes.xaxis.set_ticklabels([])  # ticks 제거
  ax[idx].axes.yaxis.set_ticklabels([])
  ax[idx].bar(range(len(df)), key_counts[idx], width=1.0)
  ax[idx].set_ylabel(labels[idx])
  ax[idx].set_xlim(0, len(df))                 # x축 범위
  ax[idx].set_ylim(0, 1)                       # y축 범위

plt.show()

# 64가지 경우의 수에 대한 분포

# init dict
key_counts = {}
for num in range(64):
  s = bin(num)[2:]
  s = s.zfill(6)
  key_counts[s] = 0

for input in df["input"].to_list():
  key_counts[input] += 1

print(df["input"].value_counts())

plt.rcParams['figure.figsize'] = [16, 6]
fig = plt.figure()
plt.bar(key_counts.keys(), key_counts.values())
plt.xticks(rotation=270)
plt.show()

print("")
```
